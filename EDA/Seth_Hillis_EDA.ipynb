```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
df = pd.read_csv("../Students_Grading_Dataset_Biased.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
cols = ['Attendance (%)', 'Midterm_Score', 'Final_Score',
       'Assignments_Avg', 'Quizzes_Avg', 'Participation_Score',
       'Projects_Score', 'Total_Score', 'Study_Hours_per_Week',
       'Extracurricular_Activities', 'Internet_Access_at_Home',
        'Family_Income_Level', 'Stress_Level (1-10)',
       'Sleep_Hours_per_Night']

for column in cols:
        plt.figure()  # Create a new figure for each plot
        plt.hist(df[column],bins=50)
        plt.title(f'Histogram of {column}')
        plt.xlabel(column)
        plt.show()
```
